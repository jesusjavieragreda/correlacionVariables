```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Paso 1: Leer el archivo Excel
file_path = "Grafo_200nodos_0'03densidad.xlsx"

# Cargar los datos desde el archivo Excel en un DataFrame
df = pd.read_excel(file_path)

# Ver los primeros registros para asegurarnos de que la carga es correcta
print("Primeros registros del archivo Excel:")
print(df.head())

# Paso 2: Calcular la correlación de cada variable con "número de iteraciones"
variables_of_interest = ['centralidad de grado', 'centralidad de interconexion',
                         'centralidad de cercania', 'excentricidad']
correlation_with_iterations = {
    variable: df[variable].corr(df['numero de iteraciones']) for variable in variables_of_interest
}

# Crear un DataFrame con las correlaciones
correlation_df = pd.DataFrame.from_dict(correlation_with_iterations, orient='index', columns=['Correlación con número de iteraciones'])

# Paso 3: Imprimir las correlaciones específicas
print("\nCorrelación de cada variable con 'número de iteraciones':")
print(correlation_df)

# Paso 4: Visualizar las correlaciones con un gráfico de barras
plt.figure(figsize=(8, 5))
sns.barplot(x=correlation_df.index, y=correlation_df['Correlación con número de iteraciones'], palette='coolwarm')
plt.title('Correlación de Variables con Número de Iteraciones')
plt.ylabel('Correlación')
plt.xlabel('Variables')
plt.xticks(rotation=45)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Línea base en 0
plt.show()
```
